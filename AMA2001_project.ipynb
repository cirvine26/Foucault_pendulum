```
import numpy as np
import matplotlib.pyplot as plt 


'''Demonstrating Elliptical path of Foucault Pendulum'''
#Initialising values
lam_d=54
lam=lam_d*(np.pi/180)
t=np.linspace(0,(24/np.sin(lam)),100)
omega=2*np.pi/24

x=np.cos(omega*np.sin(lam)*t)
y=-np.sin(omega*np.sin(lam)*t)

plt.figure(figsize=(7.5,7.5))
plt.plot(x,y)
plt.xlabel('x trace after one revolution')
plt.ylabel('y trace after one revolution')
```

```
np.seterr(divide='ignore', invalid='ignore')

'''Demonstrating how period varies with latitude in the Foucault Pendulum'''

#Initialising Values
lam=np.arange(0,2*np.pi,0.001)
T_E=np.zeros(len(lam))
T_E+=24
sin_lam=np.absolute(np.sin(lam))
T=np.divide(T_E,sin_lam)

plt.plot((lam*(180/np.pi)),T)
plt.ylim([23.5,50])
plt.xlabel('Latitude of Pendulum (Degrees)')
plt.ylabel('Period of Precession (Hours)')
```
